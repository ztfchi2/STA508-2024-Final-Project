```python
import pandas as pd
from scipy.io import arff
```

### spam.csv

```python
# codes from: https://www.kaggle.com/code/gauravduttakiit/arff-to-csv
data = arff.loadarff("/content/drive/MyDrive/Colab Notebooks/arff to csv/dataset_44_spambase.arff")
```

```python
data = pd.DataFrame(data[0])
cat = [col for col in data.columns if data[col].dtype=="O"]
data[cat] = data[cat].apply(lambda x: x.str.decode('utf8'))
data.head()
```

```python
data.to_csv('./spam.csv',index=False)
```
